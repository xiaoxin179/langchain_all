## Q: langchain_community和langchain这个包之间的关系？

**简要回答：**

| 包名 | 作用 |
|------|------|
| `langchain` | 核心包，包含 Chain、Agent 等核心抽象 |
| `langchain-core` | 底层依赖，提供 BaseChatModel、Runnable 等核心接口 |
| `langchain-community` | 第三方工具集成集合（如 DeepSeek、Chroma 等适配器） |

层层依赖：`langchain` → `langchain-core` 提供基础，`langchain-community` 扩展第三方集成。
## invoke的作用是什么？
可以理解为封装了调用的过程

```python
# 与大模型的初次交互
import os
import sys
sys.path.insert(0, '..')
from langchain_chat.config import get_deepseek_key

# 设置DeepSeek API密钥
os.environ["DEEPSEEK_API_KEY"] = get_deepseek_key()

from langchain.chat_models import init_chat_model
from langchain_core.messages import HumanMessage, SystemMessage, content

# 创建访问DeepSeek的Model
model = init_chat_model("deepseek-chat", model_provider="deepseek")

# 定义消息
messages = [
    SystemMessage(content="翻译以下内容为中文，其他的多余的不要说"),
    HumanMessage(content="how are you?"),
]
print("定义消息的messages:",messages)
user_message="你是谁？"


# 调用模型并打印结果
response = model.invoke(messages)
print("--------------------------------")
print(response.content)
```

```python
# 使用open格式的api
import os
import sys
sys.path.insert(0, '..')
from langchain_chat.config import get_deepseek_key
# 设置DeepSeek API密钥
os.environ["OPENAI_API_KEY"] = get_deepseek_key()

from langchain_openai import ChatOpenAI
llm = ChatOpenAI(
    model="deepseek-chat",
    base_url="https://api.deepseek.com",
)
response=llm.invoke([
    SystemMessage("翻译为日语"),
    HumanMessage("Hello, how are you?")
])
print(response.content)
```

```python
# langchain流式输出
from langchain_openai import ChatOpenAI
llm = ChatOpenAI(
    model="deepseek-chat",
    base_url="https://api.deepseek.com/v1",
    openai_api_key=get_deepseek_key(),
)
stream = llm.stream(
    [SystemMessage("翻译为日语"),
     HumanMessage("你是谁？你能帮我解决什么问题？")],
    stop=["不过", "但是", "然而"]
)
print(stream)
for chunk in stream:
    print(chunk.text(),end="")
```
